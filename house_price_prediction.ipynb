```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#importing boston house price data set
house_price_dataset = sklearn.datasets.fetch_california_housing()
print(house_price_dataset)
```

```python
#loading the dataset to a pandas data frame
house_price_dataframe=pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
house_price_dataframe.head()
```

```python
house_price_dataframe['price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
#checking shape
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measure of data
house_price_dataframe.describe()
```

```python
#understanding the correlation between various feature in the dataset
#1.positive correlation or 2. negetive corelation
correlation =house_price_dataframe.corr()
```

```python
print(correlation)
```

```python
#constructing a heatmap
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square= True, fmt='.1f',annot=True, annot_kws={'size':10}, cmap='Blues')
```

```python
x=house_price_dataframe.drop(['price'],axis=1)
y=house_price_dataframe['price']
```

```python
print(x)
print(y)
```

```python
#split data into training data and test data
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
#xgboost regressor
#loading
model=XGBRegressor()
```

```python
#training the model
model.fit(x_train,y_train)
```

```python
#evaluation
#prediction on training data
#accuray for prediction of training data
training_data_prediction=model.predict(x_train)
```

```python
print(training_data_prediction)
```

```python
# r sqyared error
#to finding error
score_1=metrics.r2_score(y_train,training_data_prediction)
#mean absolute error
score_2=metrics.mean_absolute_error(y_train,training_data_prediction)
print(score_1)
print(score_2)
```

```python
#prediction on test data
test_data_prediction=model.predict(x_test)
```

```python
score_1=metrics.r2_score(y_test,test_data_prediction)
#mean absolute error
score_2=metrics.mean_absolute_error(y_test,test_data_prediction)
print(score_1)
print(score_2)
```

```python
#visualize the actual prices and predicted prices
plt.scatter(y_train,training_data_prediction)
plt.xlabel('actual price')
plt.ylabel('predicted price')
plt.title('actual vs predicted price')
plt.show()
```
